```python
import pandas as pd
import numpy as np

%matplotlib notebook
```

```python
data = pd.read_csv('census.csv')
data.head()
```

```python
data.describe()
```

```python
income_raw = data['income']
feature_raw = data.drop('income',axis = 1)
numerical = ['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
```

```python
skewed = ['capital-gain', 'capital-loss']
feature_raw[skewed] = feature_raw[skewed].apply(lambda x : np.log(x+1))
```

```python
feature_raw[skewed].boxplot()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(feature_raw[numerical])
scaled_data = scaler.transform(feature_raw[numerical])
scaled_df = pd.DataFrame(data=scaled_data,columns = numerical)
scaled_df.head()
```

```python
scaled_df.boxplot()
```

```python
scaled_df[skewed].hist(bins = 25)
```

```python
skewed = ['capital-gain', 'capital-loss']
scaled_df[skewed] = scaled_df[skewed].apply(lambda x : np.log(x+1))
scaled_df.boxplot()
```

```python
scaled_df[skewed].hist(bins = 25)
```

```python
data[skewed].hist(bins = 25)
```

```python
feature_raw[numerical] = scaled_df[numerical]
```

```python
feature_raw.head()
```

```python
features = pd.get_dummies(feature_raw)
features.head()
```

```python
income = income_raw.map({'<=50K':0,'>50K':1})
income.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    income, 
                                                    test_size = 0.2, 
                                                    random_state = 0)
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,precision_score,recall_score
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train,y_train)
```

```python
predictions = naive_bayes.predict(X_test)
```

```python
print(accuracy_score(y_test,predictions))
print(f1_score(y_test,predictions))
print(confusion_matrix(y_test,predictions))
print(precision_score(y_test,predictions))
print(recall_score(y_test,predictions))
```
